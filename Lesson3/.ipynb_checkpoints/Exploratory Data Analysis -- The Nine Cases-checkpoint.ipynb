## Case 1: y_n, x_n

In this case the traditional way is the scatter plot, and seaborn is great for doing this.

```python
fig, ax = plt.subplots(figsize = (8,6))
```

```python
# Y = Cholesterol; X = hdl
# 1.1 Traditional scatter plot with matplotlib
fig, ax = plt.subplots(figsize = (8,6))
ax.scatter(diabetes["hdl"], diabetes["Cholesterol"])
ax.set_title("Traditional Scatter MPL", size = 17)
ax.set_xlabel("hdl", size = 14)
ax.set_ylabel("Cholesterol", size = 14)
plt.show()
```

### Regplot
In the simplest invocation, it draws a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression:

```python
# Regplot seaborn
fig, ax = plt.subplots(figsize = (16,6), ncols = 2)
tipo = ["linear","lowess"]
for graph,t in zip(ax,tipo):
    titulo = "lmplot seaborn " + t
    low = False
    if t== "lowess":
        low = True
    sns.regplot(x = "hdl", y = "Cholesterol", data = diabetes, lowess=low, ax = graph)
    graph.set_title(titulo, size = 17)
    graph.set_xlabel("hdl", size = 14)
    graph.set_ylabel("Cholesterol", size = 14)

plt.show()
```

### Jointplot
This is a great way to complement a scatter plot by showing the distributions of the tow variables

```python
tipos = ["scatter", "kde", "hex","reg","resid"]
for k in tipos:
    titulo = "jointplot seaborn: " + k
    g = sns.jointplot(x = "hdl", y = "Cholesterol", data = diabetes, kind = k)
    plt.show()    
```

```python
### Getting more control with JointGrid
from scipy import stats
g = sns.JointGrid(x="hdl", y="Cholesterol", data=diabetes)
g = g.plot(sns.regplot, sns.distplot)
g.annotate(stats.pearsonr, fontsize=16)
```

## Case 2: Y_n, X_c

In this case the traditional aproach is: Boxplot, violin plots, comparison of the means (or other averages) and comparison of distributions

```python
# Boxplot and Violin plot
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (14,6))
sns.boxplot(x = "frame", y = "Cholesterol", data = diabetes, order = ["small","medium","large"], ax=ax1)
sns.violinplot(x = "frame", y = "Cholesterol", data = diabetes, order = ["small","medium","large"], ax=ax2, inner="quartile")
```

```python
#fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (14,6))
tipos = ["point", "bar", "box", "violin", "strip"]
for k in tipos:
    sns.factorplot(x = "frame", y = "Cholesterol", data = diabetes, kind=k, size = 6, order = ["small","medium","large"])
    plt.show()
```

```python
# Comparison of distributions
sns.set_context('notebook', font_scale=2)
fig = sns.FacetGrid(diabetes, hue = "frame", aspect = 3, size = 6)
fig.map(sns.kdeplot, "Cholesterol", shade = True)
fig.add_legend()
```

```python
sns.set_context('notebook', font_scale=2)
fig = sns.FacetGrid(diabetes, hue = "gender", aspect = 3, size = 6)
fig.map(sns.distplot, "hdl")
fig.add_legend()
```

## Case 3 : Y_c, X_c

```python
sns.set_context('notebook', font_scale=1)
sns.countplot(x = "gender", hue = "frame", data = diabetes)
```

```python
df2 = diabetes.groupby(['gender',"frame"]).count().unstack().fillna(0)
df2["id"].plot(kind = "bar", stacked = True)
```
